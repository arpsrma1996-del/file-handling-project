```python
def update(sales_data, product_details, product_descriptions):
    sku = input("SKU (13 chars): ").strip()
    if len(sku) != 13:
        print("Invalid SKU")
        return sales_data, product_details, product_descriptions

    nums = input("Enter 14 numbers: ").split()
    if len(nums) != 14:
        print("Need exactly 14 numbers")
        return sales_data, product_details, product_descriptions

    # Update sales data
    sales_data[sku] = list(map(int, nums))

    # Update product details
    product_details[sku] = {
        "product_name": input("Name: "),
        "brand": input("Brand: "),
        "model": input("Model: "),
        "specifications": input("Specs: "),
        "price": input("Price: "),
        "availability": input("Availability: ")
    }

    # Update product description
    product_descriptions[sku] = input("Description: ")

    print("Updated successfully!")
    return sales_data, product_details, product_descriptions
```

```python
def dump_data(sales_data, product_details, product_descriptions, MAIN):
    os.makedirs(f"{MAIN}/product_details", exist_ok=True)
    os.makedirs(f"{MAIN}/product_descriptions", exist_ok=True)

    # Write CSV
    with open(f"{MAIN}/sales_data.csv", "w", newline="") as f:
        w = csv.writer(f)
        w.writerow(["Product_SKU"] + [f"Day{i}" for i in range(1, 15)])
        for sku, days in sales_data.items():
            w.writerow([sku] + days)

    # Write JSON files
    for sku, d in product_details.items():
        with open(f"{MAIN}/product_details/details_{sku}.json", "w") as jf:
            json.dump(d, jf, indent=2)

    # Write TXT files
    for sku, text in product_descriptions.items():
        with open(f"{MAIN}/product_descriptions/description_{sku}.txt", "w") as tf:
            tf.write(text)

    print("All data saved!")
```

```python
dump_data(sales_data, product_details, product_descriptions, MAIN)
```
